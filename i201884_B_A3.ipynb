```python
#importing chess function
import chess
```

```python
#Displaying board
# b = board
def Display(b):
    print(b)
#printing the board
```

```python
# Function to print game status
def print_game_status(b):
    #if checkmate
    if b.is_checkmate():
        print("Checkmate!")
    #if Stalemate
    elif b.is_stalemate():
        print("Stalemate!")
    #if Insufficient material!
    elif b.is_insufficient_material():
        print("Insufficient material!")
    #if 75-move rule!
    elif b.is_seventyfive_moves():
        print("75-move rule!")
    #if Fivefold repetition!
    elif b.is_fivefold_repetition():
        print("Fivefold repetition!")
    #if Variant draw!
    elif b.is_variant_draw():
        print("Variant draw!")
    #if check
    elif b.is_check():
        print("Check!")
```

```python
#This function is to evaluate the moves
#b is the board
def Evaluate(b):
    #white is the white material of the board
    #calculating the sum of all the tuples
    white = sum(
        #intializing for loop and making variable i for each iteration
        #getting the piece type of the board of that location in i location
        b.piece_at(i).piece_type
        #looping and idexing i at each square 
        for i in chess.SQUARES
        if b.piece_at(i) is not None and b.piece_at(i).color == chess.WHITE
        #if the piece at index i is not nol and color of the piece is white
    )
    #black is the black material of the board
    #calculating the sum of all the tuples
    black = sum(
        #intializing for loop and making variable i for each iteration
        #getting the piece type of the board of that location in i location
        b.piece_at(i).piece_type
        #looping and idexing i at each square 
        for i in chess.SQUARES
        if b.piece_at(i) is not None and b.piece_at(i).color == chess.BLACK
        #if the piece at index i is not nol and color of the piece is black
    )
    #making a variable material which is = white material - black material
    material = white - black
    return material
    #returning material
    
```

```python
# Function to get user input for search depth and validate it
def get_user_depth():
    while True:
        try:
            search_depth = int(input("Enter AI search depth (1-3): "))
            if search_depth >= 1 and search_depth <= 3:
                return search_depth
            else:
                print("Invalid search depth. Please enter a value between 1 and 3.")
        except ValueError:
            print("Invalid input, please try again.")
```

```python
#making the best move function which is making the best move 
#parameters are board and depth
def BestMove(b, d):
    #initializating variable best_move and storing none in it
    best_move = None
    #initializating max_eval which check the worst possible score of the player is maximizing
    max_eval = float('-inf') #making it float type
    #making a for loop of each move and it should be legal
    for move in b.legal_moves:
        #pushing the moves in stack
        b.push(move)
        #initializating move_eval variable which is storing the resurcively called Minmax function
        move_eval = Minimax(b, d - 1, False, float('-inf'), float('inf'))
        #poping the values from stack
        b.pop()
        #if condition to check if move_eval is great than max_eval
        if move_eval > max_eval:
            #equailizing best_move and move
            best_move = move
            #equailizing max_eval and move_eval
            max_eval = move_eval

    return best_move
    #returning best_move
```

```python
# Function to get user input for move and validate it
def get_user_move(b):
    while True:
        move = input("\nEnter your move (using UCI notation): ")
        try:
            move_obj = chess.Move.from_uci(move)
            if move_obj in b.legal_moves:
                return move_obj
            else:
                print("\nIllegal move, please try again.\n")
        except ValueError:
            print("\nInvalid input, please try again.\n")
```

```python
#making the minimax algorithum
#b=Board d=depth a=alpha be=beta is_maximizing(boolean expression to check whether a node is maximizing or not)
def Minimax(b, d, is_maximizing, a, be):
    #making an if condition which is checking if depth = 0 or the game is over
    if d == 0 or b.is_game_over():
        #if condition if game is over
        if b.is_game_over():
            print("Game Over")
            #printing game over
        return Evaluate(b)
        #returing the evaluate function
    #checking if is maximizing is true
    if is_maximizing:
        #initializating max_eval which check the worst possible score of the player is maximizing
        max_eval = float('-inf') #making it float type
        #making a for loop of each move and it should be legal
        for move in b.legal_moves:
            #pushing the moves in stack
            b.push(move)
            #initializating eval variable which is storing the resurcively called Minmax function
            eval = Minimax(b, d - 1, False, a, be)
            #poping the values from stack
            b.pop()
            #initializating max_eval variable in which the max value between max_eval and eval is stored
            max_eval = max(max_eval, eval)
            #storing the max value between alpha and eval in alpha
            a = max(a, eval)
            #if condition to check if beta <= than alpha
            if be <= a:
                #if yes break the function
                break
        #return max_eval
        return max_eval
     #checking if is maximizing is false
    else:
        #initializating min_eval which check the worst possible score of the player is not maximizing
        min_eval = float('inf') #making it float type
        #making a for loop of each move and it should be legal
        for move in b.legal_moves:
            #pushing the moves in stack
            b.push(move)
            #initializating eval variable which is storing the resurcively called Minmax function
            eval = Minimax(b, d - 1, True, a, be)
            #poping the values from stack
            b.pop()
            #initializating min_eval variable in which the max value between max_eval and eval is stored
            min_eval = min(min_eval, eval)
            #storing the min value between beta and eval in beta
            be = min(be, eval)
            #if condition to check if beta <= than alpha
            if be <= a:
                break
                #if yes break the function
        return min_eval
        #return min_eval
```

```python
#main function
def main():
    #storing the board in b
    b = chess.Board()
    search_depth = get_user_depth()

    while not b.is_game_over():
        Display(b)
        print_game_status(b)
        if b.turn:
            move_obj = get_user_move(b)
            b.push(move_obj)
        else:
            print("\nAI is making a move\n")
            ai_move = BestMove(b, search_depth)
            b.push(ai_move)
```

```python
#if main is called, calling its function
if __name__ == "__main__":
    main()
```
